## üî• Correlation Analysis

We‚Äôll check how grades in different subjects relate to each other.  
Questions we‚Äôll answer:
- Do students who perform well in Programming also do well in OOP?
- Are mathematical and probabilistic skills correlated?
- Which subjects are independent?

A **heatmap** will help visualize these correlations.

```python
numeric_cols = [
    'Proba_Stat',               
    'Analyse_Numerique',        
    'C++',                      
    'Architecture_des_syst√®mes_√†_Microprocesseu',  
    'Base_de_Donn√©es',         
    'Technologies_Web ',       
    'Analyse_et_conception_des_Syst√®mes_dInformation'  
]
df_numeric = df[numeric_cols]
```

```python
corr_matrix = df_numeric.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(
    corr_matrix,
    annot=True,      
    cmap='coolwarm', #
    vmin=-1,         
    vmax=1,          
    linewidths=0.5,
    fmt=".2f"        
)
plt.title("Correlation Between Subjects", fontsize=16)
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Select target (what we want to predict) - let's use 'Proba_Stat' as an example
target = 'Proba_Stat'

# Select features (other subjects) - exclude non-grade columns and the target
features = [col for col in df.columns if col not in ['Student_ID', 'Last_Name', 'First_Name', target]]

# Create feature matrix (X) and target vector (y)
X = df[features]
y = df[target]
```

```python
X
```

```python
y
```

```python
# Split data into 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions on test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")
```

```python
coefficients = pd.DataFrame({
    'Feature': features,
    'Coefficient': model.coef_
}).sort_values('Coefficient', ascending=False)

print("\nFeature Coefficients:")
print(coefficients)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)  # Perfect prediction line
plt.xlabel('Actual Grades')
plt.ylabel('Predicted Grades')
plt.title(f'Actual vs Predicted {target} Grades')
plt.show()
```
